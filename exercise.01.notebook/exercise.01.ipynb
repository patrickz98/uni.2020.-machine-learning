```python
import matplotlib.pyplot as plt
import json
import math

labels = []

with open('sin-points-with-noise.json') as json_file:
    random = json.load(json_file)
    plt.plot(random["XPoints"], random["YPoints"], "o")
    labels += ["Training data points"]

    sinStuff = [math.sin(2 * math.pi * x) for x in random["XPoints"]]
    plt.plot(random["XPoints"], sinStuff)
    labels += ["Sin function"]


with open('stochastic-gradient-descent.results.json') as json_file:
    data = json.load(json_file)
    
    for plot in data:
        plt.plot(plot["XPoints"], plot["YPoints"])
        labels += [plot["Name"]]

plt.legend(labels)
plt.savefig('exercise.01.results.png', dpi=500)
```
