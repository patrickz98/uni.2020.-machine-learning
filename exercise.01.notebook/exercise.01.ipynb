```python
import matplotlib.pyplot as plt
import json

labels = []

with open('exercise.01.json') as json_file:
    random = json.load(json_file)
    plt.plot(random["XPoints"], random["YPoints"])
    labels += ["Random init"]

with open('exercise.01.sgd.json') as json_file:
    data = json.load(json_file)
    
    for plot in data:
        plt.plot(plot["XPoints"], plot["YPoints"])
        labels += [plot["Name"]]

plt.legend(labels)
plt.savefig('exercise.01.png', dpi=500)
```
