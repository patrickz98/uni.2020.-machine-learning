```python
import matplotlib.pyplot as plt
import json

labels = []

with open('model.final.json') as json_file:
    model = json.load(json_file)
    plt.plot(model["XPoints"], model["YPoints"])
    labels += ["Final Model"]


with open('model.init.json') as json_file:
    model = json.load(json_file)
    plt.plot(model["XPoints"], model["YPoints"])
    labels += ["Init Model"]


with open('trainingPoints.json') as json_file:
    data = json.load(json_file)

    xPointsGreen = []
    yPointsGreen = []
    xPointsRed = []
    yPointsRed = []

    for point in data:
        if point[ "Label" ] == 1:
            xPointsGreen += [point[ "Dimension1" ]]
            yPointsGreen += [point[ "Dimension2" ]]
        else:
            xPointsRed += [point[ "Dimension1" ]]
            yPointsRed += [point[ "Dimension2" ]]

    plt.plot(xPointsGreen, yPointsGreen, "go")
    plt.plot(xPointsRed, yPointsRed, "ro")

plt.xlim(-3, 3)
plt.ylim(-8, 4)
plt.legend(labels)
plt.savefig('exercise.02.results.png', dpi=500)
```
